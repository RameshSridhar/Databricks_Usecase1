### Creating pyspark dataframe and reading the above file

```python
file_path = "/Volumes/workspace/default/usage_metrics/mobile_os_usage.csv"

df1 = spark.read.csv(file_path, header=True, inferSchema=True)
df1.show(5)
```

### Writing the dataframe df1 data into databricks table

```python
%python
from pyspark.sql.functions import col

df1_clean = df1.toDF(*(c.replace(" ", "-").replace(",","").replace("%","")for c in df1.columns))

df1_clean.write.mode("overwrite").saveAsTable("default.mobile_os_usage")
print("Table created")
```

### Writing sql query to display the data loaded into the databricks table

```python
%python
spark.sql("select *from default.mobile_os_usage limit 10").show()
```
